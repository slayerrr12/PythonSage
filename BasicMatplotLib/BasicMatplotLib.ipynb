```python
# Your data
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]

# Plotting
plt.plot(x,y)
plt.xlabel("data")
plt.ylabel("random")
plt.title("random data")



# Customize x-axis ticks
plt.xticks([1, 2, 3, 4, 5,100])

# Customize y-axis ticks
plt.yticks([2, 3, 4, 5, 6,100])

# Display the plot
plt.show()

```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]

# Plotting
lines = plt.plot(x, y)

# 'lines' is a list of Line2D objects
print(lines)
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 4, 5, 6]
y2 = [1, 2, 1, 2, 1]

# Plotting with labels
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```
