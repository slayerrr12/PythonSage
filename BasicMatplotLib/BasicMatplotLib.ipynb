## Barplot Examples with parameters

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 30, 20]

# Creating a bar plot
plt.bar(categories,values, color="red",edgecolor='black')
# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')

# Adding grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data for the scatter plot
np.random.seed(42)
x = np.random.rand(50)
y = 3 * x + 2 + 0.1 * np.random.randn(50)  # Linear relationship with some noise

# Creating a scatter plot with custom parameters
plt.scatter(
    x,
    y,
    s=90,               # Marker size
    c='blue',           # Marker color
    marker='o',         # Marker style
    cmap='viridis',     # Colormap
    alpha=0.7,          # Transparency
    edgecolors='black', # Edge color of markers
    linewidths=1.2,     # Linewidth of marker edges
    label='Data Points'  # Label for the legend
)

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Adding a colorbar
plt.colorbar(label='Color Intensity')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```
